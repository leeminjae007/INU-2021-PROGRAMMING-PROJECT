```python
import numpy as np
import pandas as pd

total_data = pd.DataFrame()

for month_index in range(1,13):
  month_data = pd.read_excel('/content/drive/MyDrive/ProgramHW/2020/2020년 ' + str(month_index) + '월.xlsx')
  total_data = total_data.append(month_data, ignore_index = True)

print(total_data.shape)
```

```python
busan_data = total_data[total_data['지역'].str.contains('부산')]
busan_data.reset_index(drop = True, inplace = True)
```

```python
busan_data.to_csv('/content/drive/MyDrive/ProgramHW/busan_data.csv', index = False)
```
