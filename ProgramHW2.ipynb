```python
import pandas as pd
import numpy as np

busan_data = pd.read_csv('/content/drive/MyDrive/ProgramHW/busan_data.csv')
```

```python
local_observatory = set(busan_data['측정소코드'].values)
local_observatory = list(local_observatory)

print("총 관측소의 개수는 " + str(len(local_observatory)) + "개 이며 누락된 데이터 개수는 다음과 같습니다.")
print(busan_data.isnull().sum())
```

```python
SO2_Color_list = []
CO_Color_list = []
O3_Color_list = []
NO2_Color_list = []
PM10_Color_list = []
PM2dot5_Color_list = []

for index in range(0, len(busan_data)):
  if np.isnan(busan_data.loc[index, 'SO2']) == True:
    SO2_Color_list.append('None')
  else:
    if float(busan_data.loc[index, 'SO2']) <= 0.02:
      SO2_Color_list.append('Blue')
    elif 0.02 < float(busan_data.loc[index, 'SO2']) <= 0.05:
      SO2_Color_list.append('Green')
    elif 0.05 < float(busan_data.loc[index, 'SO2']) <= 0.15:
      SO2_Color_list.append('Yellow')
    else:
      SO2_Color_list.append('Red')

for index in range(0, len(busan_data)):
  if np.isnan(busan_data.loc[index, 'CO']) == True:
    CO_Color_list.append('None')
  else:
    if float(busan_data.loc[index, 'CO']) <= 2:
      CO_Color_list.append('Blue')
    elif 2 < float(busan_data.loc[index, 'CO']) <= 9:
      CO_Color_list.append('Green')
    elif 9 < float(busan_data.loc[index, 'CO']) <= 15:
      CO_Color_list.append('Yellow')
    else:
      CO_Color_list.append('Red')

for index in range(0, len(busan_data)):
  if np.isnan(busan_data.loc[index, 'O3']) == True:
    O3_Color_list.append('None')
  else:
    if float(busan_data.loc[index, 'O3']) <= 0.03:
      O3_Color_list.append('Blue')
    elif 0.003 < float(busan_data.loc[index, 'O3']) <= 0.09:
      O3_Color_list.append('Green')
    elif 0.09 < float(busan_data.loc[index, 'O3']) <= 0.15:
      O3_Color_list.append('Yellow')
    else:
      O3_Color_list.append('Red')

for index in range(0, len(busan_data)):
  if np.isnan(busan_data.loc[index, 'NO2']) == True:
    NO2_Color_list.append('None')
  else:
    if float(busan_data.loc[index, 'NO2']) <= 0.03:
      NO2_Color_list.append('Blue')
    elif 0.03 < float(busan_data.loc[index, 'NO2']) <= 0.06:
      NO2_Color_list.append('Green')
    elif 0.06 < float(busan_data.loc[index, 'NO2']) <= 0.2:
      NO2_Color_list.append('Yellow')
    else:
      NO2_Color_list.append('Red')


for index in range(0, len(busan_data)):
  if np.isnan(busan_data.loc[index, 'PM10']) == True:
    PM10_Color_list.append('None')
  else:
    if float(busan_data.loc[index, 'PM10']) <= 30:
      PM10_Color_list.append('Blue')
    elif 30 < float(busan_data.loc[index, 'PM10']) <= 80:
      PM10_Color_list.append('Green')
    elif 80 < float(busan_data.loc[index, 'PM10']) <= 150:
      PM10_Color_list.append('Yellow')
    else:
      PM10_Color_list.append('Red')


for index in range(0, len(busan_data)):
  if np.isnan(busan_data.loc[index, 'PM25']) == :
    PM2dot5_Color_list.append('None')
  else:
    if float(busan_data.loc[index, 'PM25']) <= 15:
      PM2dot5_Color_list.append('Blue')
    elif 15 < float(busan_data.loc[index, 'PM25']) <= 35:
      PM2dot5_Color_list.append('Green')
    elif 35 < float(busan_data.loc[index, 'PM25']) <= 75:
      PM2dot5_Color_list.append('Yellow')
    else:
      PM2dot5_Color_list.append('Red')
```

```python
busan_data['SO2Color'] = SO2_Color_list
busan_data['COColor'] = CO_Color_list
busan_data['O3Color'] = O3_Color_list
busan_data['NO2Color'] = NO2_Color_list
busan_data['PM10Color'] = PM10_Color_list
busan_data['PM25Color'] = PM2dot5_Color_list

busan_data.to_csv('/content/drive/MyDrive/ProgramHW/busan_edit_data.csv', index = False)
```
