```python
import pandas as pd
import numpy as np
import googlemaps
import folium
```

```python
busan_data = pd.read_csv('/content/drive/MyDrive/ProgramHW/busan_edit_data.csv')
```

```python
observatory_name = set(busan_data['측정소명'].values)
observatory_name = list(observatory_name)
observatory_num = set(busan_data['측정소코드'].values)
observatory_num = list(observatory_num)
observatory_lat = []
observatory_long = []

for observatory_index in observatory_name:
  gmaps = googlemaps.Client(key = 'AIzaSyDDh_1phUtXNtyzpqU17RuKth2WAl2eAQY')
  geocode_result = gmaps.geocode('부산 ' + str(observatory_index), language = 'ko')
  lat = geocode_result[0]['geometry']['location']['lat']
  lon = geocode_result[0]['geometry']['location']['lng']
  observatory_lat.append(lat)
  observatory_long.append(lon)

location_data = pd.DataFrame()
location_data['측정소명'] = observatory_name
location_data['측정소코드'] = observatory_num
location_data['lat'] = observatory_lat
location_data['long'] = observatory_long
print(location_data)
```

```python
m = folium.Map(location = [35.200124528698116, 129.07491267021746], zoom_start = 11)

for index in range (0, len(location_data)):
  folium.Marker(
      location = [location_data.loc[index, 'lat'], location_data.loc[index, 'long']],
      popup = str(location_data.loc[index, '측정소코드']),
      icon = folium.Icon(color = 'red', icon = 'star')
  ).add_to(m)

m
```
